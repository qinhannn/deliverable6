```python
#1. Read and populate file
df_auto = pd.read_csv("auto-mpg.data", sep = '\s+')
df_auto.head()
```

```python
#2. Give df headers
headers = ["mpg", "cylinders", "displacement", "horsepower", "weight", "acceleration", "model_year", "origin", "car_name"]
df_auto.columns = headers 
df_auto.head()
```
